```python
import pandas as pd
data = pd.read_csv('MLAPBD.csv')
data.drop('Unnamed: 0', axis= 1, inplace= True)
data
```

```python
data_new = data.apply(lambda x: pd.factorize(x)[0])
```

```python
data_new['satuanoutcome']= data_new['satuanoutcome'].replace({-1 : 137})
data_new['satuanoutcome'].unique()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
x =  MinMaxScaler().fit_transform(data_new)
y = data['nilaianggaran']
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.01, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_case_tree = DecisionTreeRegressor (max_depth=40,min_samples_split= 5)

model_case_tree.fit(X_train,y_train)


```

```python
nilai_anggaran = model_case_tree.predict(X_test)
print(nilai_anggaran)
semua_nilai_anggaran= model_case_tree.predict(x)
print(semua_nilai_anggaran)
```

```python
print(round(mean_absolute_error(y_test,nilai_anggaran)))  #y_test x_test
```

```python
print(round(mean_absolute_error(y_train, model_case_tree.predict(X_train)))) 
```

```python
print("Training Accuracy: {0:.2f}".format(model_case_tree.score(X_train, y_train)))
print("Testing Accuracy: {0:.2f}".format(model_case_tree.score(X_test, y_test)))
```
